```python
import pandas as pd
```

```python
df=pd.read_csv("house_price_regression_dataset.csv")
```

```python
df
```

```python
df.dtypes
```

```python
df["House_Price"]=df["House_Price"].astype(int)
```

```python
df.dtypes
```

```python
y=df["House_Price"]
```

```python
y
```

```python
# x=df[["Square_Footage","Num_Bedrooms",'Year_Built',"Lot_Size","Garage_Size","Neighborhood_Quality"]]

x=df.drop("House_Price",axis=1)
```

```python
x
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
xscaled=scaler.fit_transform(x)
```

```python
xscaled
```

# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(xscaled,y,test_size=0.2,random_state=32)
```

```python
x.shape
```

```python
xtrain.shape
```

```python
xtest.shape
```

# Model building

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
```

```python
ypred
```

```python
ytest
```

```python
model.intercept_
```

```python
model.coef_
```

```python
residuals=ytest-ypred
```

```python
residuals
```

# Save the model

```python
import joblib
```

```python
joblib.dump(model,'modelhouse.pkl')
```

```python
joblib.dump(scaler,"scalerhouse.pkl")
```
